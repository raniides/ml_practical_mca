```python
#KMeans with PCA
# Selecting relevant features/columns
features = drybean.drop(columns=['Class'])

# Scale the features
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# Apply PCA
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca_features = pca.fit_transform(scaled_features)

# Fit KMeans to the PCA-transformed features
kmeans = KMeans(n_clusters=4)
labels = kmeans.fit_predict(pca_features)

centroids= kmeans.cluster_centers_
print(centroids)

# Plot the clusters
plt.figure(figsize=(10, 6))

#plt.scatter(pca_features[label == 0, 0], pca_features[label == 0, 1], color='red', label='Cluster 0')
#plt.scatter(pca_features[label == 1, 0], pca_features[label == 1, 1], color='black', label='Cluster 1')
#plt.scatter(pca_features[label == 2, 0], pca_features[label == 2, 1], color='pink', label='Cluster 2')
#plt.scatter(pca_features[label == 3, 0], pca_features[label == 3, 1], color='yellow', label='Cluster 3')

plt.scatter(pca_features[:, 0], pca_features[:, 1], c=labels.astype(float), s=50, alpha=0.5)
plt.scatter(centroids[:,0], centroids[:,1], c= "red",marker="x", s=50)

#plt.scatter(centers[:,0], centers[:,1], marker="x", color='g')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('KMeans Clustering with PCA')
plt.legend()
plt.show()
```
