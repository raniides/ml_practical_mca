```python
#dividing the dataset into the train and test set
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

df.shape
```

```python
ss = StandardScaler()
X_train =ss.fit_transform(X_train)
X_test = ss.transform(X_test)

X_train
```

```python
#applying logistic Regression
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train, y_train)

#implemented through Logistic Regression
y_pred = lr.predict(X_test)
y_pred

#array containing the output
y_test
```

```python
from sklearn.metrics import accuracy_score
lr_acc=accuracy_score(y_test, y_pred)
lr_acc
```

```python
from sklearn.metrics import precision_score
lr_pre=precision_score(y_test, y_pred, average="macro")
lr_pre
```

```python
from sklearn.metrics import recall_score
lr_rec=recall_score(y_test, y_pred, average="macro")
lr_rec
```

```python
from sklearn.metrics import f1_score
lr_f1=f1_score(y_test, y_pred, average="macro")
lr_f1
```

```python
tempResults = pd.DataFrame({'Algorithm':['Logistic Regression Method'], 'Accuracy':[lr_acc], 'precision':[lr_pre], 'recall':[lr_rec], 'f1_score':[lr_f1]})
results=pd.DataFrame()
results = pd.concat( [results, tempResults])
results = results[['Algorithm','Accuracy','precision','recall','f1_score']]
results
```

```python
# applying Decision Tree Classifier
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
dtc.fit(X_train, y_train)

y_pred = dtc.predict(X_test)
y_pred

print(accuracy_score(y_test,y_pred))
dtc_acc=accuracy_score(y_test, y_pred)

from sklearn.metrics import precision_score
dtc_pre=precision_score(y_test, y_pred, average="macro")
dtc_pre

from sklearn.metrics import recall_score
dtc_rec=recall_score(y_test, y_pred, average="macro")
dtc_rec

from sklearn.metrics import f1_score
dtc_f1=f1_score(y_test, y_pred, average="macro")
dtc_f1

tempResults = pd.DataFrame({'Algorithm':['Decision Tree Classifier Method'], 'Accuracy':[dtc_acc], 'precision':[dtc_pre], 'recall':[dtc_rec], 'f1_score':[dtc_f1]})
results=pd.DataFrame()
results = pd.concat( [results, tempResults])
results = results[['Algorithm','Accuracy','precision','recall','f1_score']]
results
```

```python
#applying Random Forest Classifier
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(X_train, y_train)

y_pred = rfc.predict(X_test)
y_pred

print(accuracy_score(y_test,y_pred))
rfc_acc=accuracy_score(y_test, y_pred)

from sklearn.metrics import precision_score
rfc_pre=precision_score(y_test, y_pred, average="macro")
rfc_pre

from sklearn.metrics import recall_score
rfc_rec=recall_score(y_test, y_pred, average="macro")
rfc_rec

from sklearn.metrics import f1_score
rfc_f1=f1_score(y_test, y_pred, average="macro")
rfc_f1

tempResults = pd.DataFrame({'Algorithm': ['Random Forest Classifier Method'],'Accuracy': [rfc_acc], 'precision':[rfc_pre], 'recall':[rfc_rec], 'f1_score':[rfc_f1]})

results = pd.concat( [results, tempResults])
results = results[['Algorithm','Accuracy','precision','recall','f1_score']]
results
```

```python
#applying Support Vector Machine
from sklearn import svm
svc = svm.SVC()
svc.fit(X_train, y_train)

y_pred = svc.predict(X_test)
y_pred

from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred))
svc_acc=accuracy_score(y_test, y_pred)

from sklearn.metrics import precision_score
svc_pre=precision_score(y_test, y_pred, average="macro")
svc_pre

from sklearn.metrics import recall_score
svc_rec=recall_score(y_test, y_pred, average="macro")
svc_rec

from sklearn.metrics import f1_score
svc_f1=f1_score(y_test, y_pred, average="macro")
svc_f1

tempResults = pd.DataFrame({'Algorithm': ['Support Vector Classifier Method'],'Accuracy': [svc_acc], 'precision':[svc_pre], 'recall':[svc_rec], 'f1_score':[svc_f1]})

results = pd.concat( [results, tempResults])
results = results[['Algorithm','Accuracy','precision','recall','f1_score']]
results
```
