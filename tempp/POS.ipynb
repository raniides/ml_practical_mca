```python
import numpy as np
import pandas as pd
import nltk 
from nltk.tokenize import word_tokenize
from nltk import pos_tag
from nltk.tag import hmm
```

```python
df=pd.read_csv("hindi_ner.csv")
```

```python
df
```

```python
x = df.iloc[:, 0]
y = df.iloc[:, 1]
```

```python
train= []
for i in range(len(x)):
    train.append((x[i], y[i]))
```

```python
train
```

```python
trainer = hmm.HiddenMarkovModelTrainer()
```

```python
tagger = trainer.train_supervised([train])
```

```python
test_sentence ="वह घर पर पढ़ रही है।"
```

```python
tagged_sentence = tagger.tag(test_sentence.split())
print(tagged_sentence)
```
